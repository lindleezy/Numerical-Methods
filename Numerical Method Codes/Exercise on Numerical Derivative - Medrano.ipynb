```python
import numpy as np
import matplotlib.pyplot as plt

def ForwardDiff(function, t, h):
    f = function
    return (f(t + h) - f(t)) / h

def CentralDiff(function, t, h):
    f = function
    return (f(t + h) - f(t - h)) / (2 * h)

def ExtrapolatedDiff(function, t, h):
    f = function
    a = 8 * (f(t + h/4) - f(t - h/4))
    b = f(t + h/2) - f(t - h/2)
    return (a - b) / (3 * h)
```

Given two functions:

$$F_{1}(\text{x}) = \cos(t)$$
<br>
$$F_{2}(\text{x}) = e^{t}$$

```python
def func1(t):
    return np.cos(t)

def func2(t):
    return np.exp(t)
```

## Evaluating the derivative of $\cos\left(t\right)$

```python
vals = []
h_vals = np.linspace(np.finfo(float).eps, 1, 1000)
t_vals = [0.1, 1, 100]

plt.figure(dpi=100)
for i in t_vals:
    vals = []
    errors = []
    true_val = -np.sin(i)
    for h in h_vals:
        val = ForwardDiff(function=func1, t=i, h=h)
        vals.append(val)
        errors.append(abs((val - true_val) / true_val))
    plt.loglog(h_vals, errors, label=f'$t = {i}$')
    plt.legend(loc='best')


plt.title(r'log-log Graph for $\varepsilon$ vs. $h$ using the Forward-Difference method')
plt.xlabel('$h$')
plt.ylabel('$\epsilon$')
plt.show()
```

### See if you can identify regions where algorithmic (series truncation) error dominates at large \(h\) and round-off error at small \(h\) in your plot. Do the slopes agree with our model’s predictions?


It is observed that all plots have similar trend. for a very small value $h$, the error is very large, this error is known as `round-off error` which happens when dealing with small values, it is basically caused by approximate representation of numbers. As $h$ increases, the error obtained decreases until a point where it blows up. This new error is known as `truncation error`, which is caused by truncating a mathematical procedure. The slopes therefore agree with the model's prediction.
